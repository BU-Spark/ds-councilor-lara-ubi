## question i'm answering: Are the parents in the families under the poverty line Married? Divorced? Or single parent?

dataset link: https://data.census.gov/table?q=Marital+Status+and+Marital+History&t=Income+and+Poverty&g=060XX00US2502507000

```python
import csv

# Read the CSV file
with open('maritalStatus.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))

# Find the indices of the relevant columns
total_population_idx = data[0].index("Boston city, Suffolk County, Massachusetts!!Total!!Estimate")
married_idx = data[0].index("Boston city, Suffolk County, Massachusetts!!Now married (except separated)!!Estimate")
widowed_idx = data[0].index("Boston city, Suffolk County, Massachusetts!!Widowed!!Estimate")
divorced_idx = data[0].index("Boston city, Suffolk County, Massachusetts!!Divorced!!Estimate")
separated_idx = data[0].index("Boston city, Suffolk County, Massachusetts!!Separated!!Estimate")
never_married_idx = data[0].index("Boston city, Suffolk County, Massachusetts!!Never married!!Estimate")

def str_to_float(s):
    return float(s.replace(",", "").replace("%", ""))

# Get the relevant data for the total population 15 years and over
total_population = str_to_float(data[1][total_population_idx])
married = data[1][married_idx]
widowed = data[1][widowed_idx]
divorced = data[1][divorced_idx]
separated = data[1][separated_idx]
never_married = data[1][never_married_idx]

# Print the results
print(f"Percentage of married people: {married}")
print(f"Percentage of widowed people: {widowed}")
print(f"Percentage of divorced people: {divorced}")
print(f"Percentage of separated people: {separated}")
print(f"Percentage of never married people: {never_married}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Boston-Poverty-Data.csv
poverty_data_df = pd.read_csv('Boston-Poverty-Data.csv')

# Get the median income for married-couple families and nonfamily households
median_income_married_couple_families = float(poverty_data_df.iloc[11]['Boston city, Massachusetts!!Married-couple families!!Estimate'].replace(',', ''))
median_income_nonfamily_households = float(poverty_data_df.iloc[11]['Boston city, Massachusetts!!Nonfamily households!!Estimate'].replace(',', ''))

# Create a bar chart to compare the median incomes
fig, ax = plt.subplots()
ax.bar(['Married-couple families', 'Nonfamily households'], [median_income_married_couple_families, median_income_nonfamily_households])
ax.set_ylabel('Median income ($)')
ax.set_title('Comparison of median income for married-couple families and nonfamily households')

# Print the comparison statement
if median_income_married_couple_families > median_income_nonfamily_households:
    comparison_statement = f"Married-couple families tend to have higher income (${median_income_married_couple_families:,.0f}) than nonfamily households (${median_income_nonfamily_households:,.0f})."
else:
    comparison_statement = f"Nonfamily households tend to have higher income (${median_income_nonfamily_households:,.0f}) than married-couple families (${median_income_married_couple_families:,.0f})."
print(comparison_statement)

plt.show()
```
