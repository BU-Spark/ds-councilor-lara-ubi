```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('Q7.csv')
```

```python
data
```

```python
df = data
estimate_cols = [col for col in df.columns if 'Estimate' in col]

# calculate the sum of the estimate columns
df['Sum of Estimates'] = df[estimate_cols].sum(axis=1)

# drop the margin of error columns
df = df.loc[:, ~df.columns.str.contains('Margin of Error')]
```

```python
df = df.loc[1:]

ax = plt.subplot()
ax.bar(df['Label (Grouping)'], df['Sum of Estimates'])
ax.set_xticklabels(df['Label (Grouping)'], rotation=90)
ax.set_xlabel('Income Group')
ax.set_ylabel('Sum of Families')
ax.set_title('Total Families by Income Group')

# add labels to the bars
for i, val in enumerate(df['Sum of Estimates']):
    ax.text(i, val, str(val), ha='center', va='bottom')

plt.show()
```

In this text, we can see that for the ratio of families under poverty line, we can see that the ratio under .50 contains 11280 families, while .50 to .74 only contains 5752, just nearly half of the ratio under .50. And the ratio .75 to .99 contains 6470 families.
