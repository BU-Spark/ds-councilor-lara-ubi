```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel('thresh22.xlsx')
```

```python
data
```

```python
# Extract the rows corresponding to "Under 65 years" and "65 years and over"
data_Size1 = data.iloc[[4, 5]]

# Plot the resulting DataFrame
ax = data_Size1.plot(kind='bar', figsize=(8, 6), legend=False)

# Set the axis labels and title
ax.set_xlabel('Number of people in a family')
ax.set_ylabel('Poverty Threshold (USD)')
ax.set_title('Poverty Threshold by Family Size for Different Age Groups')

# Specify the legend labels
ax.set_xticklabels(['Under 65 years', '65 years and over'])

plt.show()
```

```python
data_1 = data.iloc[3:]
```

```python
data1 = data_1.drop(['Table with row headings in column A and column headings in rows 5 to 6.'], axis=1)
```

```python
data1.fillna(0)
```

```python
data1=pd.DataFrame(data1,dtype=np.float)
```

```python
data1.iloc[1] = data1.iloc[1].add(data1.iloc[2])/2
data1.iloc[5] = data1.iloc[5].add(data1.iloc[6])/2
```

```python
data1
```

```python
rows_to_sum = [1, 5, 8, 9, 10, 11, 12, 13, 14]  # List of row numbers to sum
summed_row = data1.iloc[rows_to_sum].mean(axis = 1)  # Sum the selected rows
```

```python
print(f"Mean of rows {rows_to_sum}:")
print(summed_row)
```

```python
fig, ax = plt.subplots()
summed_row.T.plot(kind='bar', ax=ax)

ax.set_xlabel('Number of people in a family')
ax.set_ylabel('The Average of Poverty Threshold (USD)')
ax.set_title('Poverty Threshold by Family Size')

ax.set_xticklabels(['1', '2', '3', '4', '5', '6', '7', '8', '9'])

plt.show()
```
