```python
import pandas as pd

data_df = pd.read_excel('./Comparision_SNAP_years.xlsx', sheet_name="Sheet1")

```

```python
data_df.dropna(axis='columns')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

plt.plot(data_df["Year"], 100*data_df["Households-Receiving SNAP"])
plt.ylim(10, 30)
# add labels and title to the plot
plt.xlabel("Year")
plt.ylabel("% of Households Receiving SNAP")
plt.title("Trend of Households Receiving SNAP over the Years")

# show the plot
plt.show()
```
